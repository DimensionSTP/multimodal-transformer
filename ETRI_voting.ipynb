```python
import numpy as np
import pandas as pd
from sklearn import metrics
from sklearn.metrics import f1_score
```

```python
audio = np.load('./npy/pred_hubert_large_whole_1.npy')
text = np.load('./npy/pred_roberta_large_whole_1.npy')
test = pd.read_pickle('path_test.pkl')
```

```python
test = test.dropna()
```

```python
label = test.emotion.values
```

```python
label
```

```python
label.shape
```

```python
audio_result = audio
audio_pred = np.argmax(audio_result, 1)
audio_f1 = f1_score(label, audio_pred, average='macro')
audio_f1
```

```python
text_result = text
text_pred = np.argmax(text_result, 1)
text_f1 = f1_score(label, text_pred, average='macro')
text_f1
```

```python
result = audio*0.33 + text*0.67
pred = np.argmax(result, 1)
pred
```

```python
combined_f1 = f1_score(label, pred, average='macro')
combined_f1
```
